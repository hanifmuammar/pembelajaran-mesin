TF-IDF

```python
#tfidf adalah metode yang digunakan untuk mengukur seberapa penting suatu kata dalam suatu dokumen
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer() #membuat objek tfidf
```

```python
X = vectorizer.fit_transform(dataset_feature)
```

```python
print(X.shape)
```

```python
features = vectorizer.get_feature_names_out() #mengambil fitur dari tfidf yang sudah dihitung
```

```python
features #menampilkan fitur
```

```python
idfValues = vectorizer.idf_ #mengambil nilai idf dari tfidf 
```

```python
d = dict(zip(features, 9 - idfValues)) #membuat dictionary yang berisi fitur dan nilai idfnya 

#(zip digunakan untuk menggabungkan dua list)
```

```python
sortedDict = sorted(d.items(), key = lambda d: d[1], reverse = True) #mengurutkan dictionary berdasarkan nilai idfnya dari yang terbesar

#key = lambda d: d[1] digunakan untuk mengurutkan berdasarkan nilai idfnya
```

```python
for i in range(3867):
    print(sortedDict[i]) #menampilkan fitur dengan nilai idf terbesar 
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plot
```

```python
def PlotWordCloud(frequency): #fungsi untuk membuat wordcloud
    worcloudPlot = WordCloud(background_color="white", width=1500, height=1000) #membuat objek wordcloud
    worcloudPlot.generate_from_frequencies(frequencies=frequency) #membuat wordcloud dari frekuensi kata
    plot.figure(figsize=(15,10)) #membuat figure
    plot.imshow(worcloudPlot, interpolation="bilinear")
    plot.axis("off")
    plot.show()
```

```python
PlotWordCloud(d) #menampilkan wordcloud dari fitur dan nilai idfnya
```
