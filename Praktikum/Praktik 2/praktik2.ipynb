```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # digunakan untuk membuat grafik
import seaborn as sns # digunakan untuk membuat grafik
```

```python
# visualisasi pie chart
players = ('Ronaldo', 'Messi', 'Bellingham', 'Haaland', 'Mbappe')
votes = (12, 11, 4, 8, 7)
plt.pie(
    votes, # data
    labels=players, # label
)
plt.show() # menampilkan grafik
```

```python
# melihat persen kontribusi pada pie chart
players = ('Ronaldo', 'Messi', 'Bellingham', 'Haaland', 'Mbappe')
votes = (12, 11, 4, 8, 7)
plt.pie(
    votes,
    labels=players,
    autopct='%1.1f%%', # menampilkan persen
)
plt.show()
```

```python
players = ('Ronaldo', 'Messi', 'Bellingham', 'Haaland', 'Mbappe')
votes = (12, 11, 4, 8, 7)
colors = ('#8B7842', '#FFF2DC', '#81C572', '#E89F0D', '#D53999')
plt.pie(
    votes,
    labels=players,
    autopct='%1.1f%%', # menampilkan persen
    colors=colors,
)
plt.show()
```
